```python
def gd(points, learning_rate, num_iterations):
    m = [0 for i in range(points.shape[1]-1)]
    c = 0
    for i in range(num_iterations):
        if i ==  1235:
            print('True')
            learning_rate = learning_rate/30 
        
        m, c = step_gradient(points, learning_rate, m, c)
        print(i, " Cost: ", cost(points, m, c))
    return m, c
```

```python
def step_gradient(points, learning_rate, m, c):
    
    num_features = points.shape[1]-1

    
    m_slope = [0 for i in range(num_features)]
    c_slope = 0
    M = points.shape[0]   # number of training examples
    for i in range(M):
        x = points[i, 0:num_features]
        y = points[i, num_features:num_features+1]
        #print(x[0],y[0])
        s=0
        for j in range(num_features):
            s+=m[j]*x[j]
        for j in range(num_features):
            m_slope[j] += (-2/M)* (y[0] - s )*x[j]
        c_slope=(-2/M)* (y[0] - s )*1
#       print(m_slope)
       
    new_m = m - np.dot(learning_rate,m_slope)
    new_c = c-learning_rate*c_slope
    return new_m, new_c
```

```python
def cost(points, m, c):
    total_cost = 0
    
    num_features = points.shape[1]-1
    m=np.array(m)
    M = points.shape[0]
    for i in range(M):
        x = points[i, 0:num_features]
        y = points[i, num_features:num_features+1]
        
        
        total_cost += (1/M)*((y[0] - np.sum(m*x)-c)**2)
        
    return total_cost
```

```python
def predict(data,m, c):

    predicted = []

    for i in range(data.shape[0]):
        predicted.append(np.sum(m*data[i])+c)
    
    print(len(predicted), predicted[:20]) 
    
    return predicted
```
